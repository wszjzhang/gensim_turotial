```python
import numpy as np
```

```python
import gensim, logging
from gensim.models import Word2Vec
```

```python
model = Word2Vec.load_word2vec_format(
    '../word2vec/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
sentence_1 = "how are you"
sentence_2 = "where are you"
sentence_3 = "how is going"
```

```python
def sentence_vec(sentence):
    words = sentence.split()
    sv = model[words[0]]
    for token in words[1:]:
        sv += model[token]

    return sv
```

```python
np.dot(sentence_vec(sentence_1), sentence_vec(sentence_2))
```

```python
np.dot(sentence_vec(sentence_1), sentence_vec(sentence_3))
```

```python
np.dot(sentence_vec(sentence_3), sentence_vec(sentence_2))
```

```python
sentence_4 = "hello"
```

```python
np.dot(sentence_vec(sentence_3), sentence_vec(sentence_4))
```
